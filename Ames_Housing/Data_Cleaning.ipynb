# Data Cleaning
This notebook handles structural missing values and casts types. It processes both `train.csv` and `test.csv` without causing any data leakage.

```python
import pandas as pd

# Load the raw splits
train_df = pd.read_csv('Ames_Housing/datasets/train.csv')
test_df = pd.read_csv('Ames_Housing/datasets/test.csv')
```

```python
def structural_cleaning(df):
    # 1. Numerical Pseudo-Missing (Fill with 0)
    num_zero_cols = [
        'Garage Area', 'Garage Cars', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 
        'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Mas Vnr Area'
    ]
    df[num_zero_cols] = df[num_zero_cols].fillna(0)
    
    # 2. Categorical Pseudo-Missing (Fill with "None")
    cat_none_cols = [
        'Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Mas Vnr Type', 
        'Fireplace Qu', 'Garage Qual', 'Garage Finish', 'Garage Cond', 
        'Garage Type', 'Bsmt Exposure', 'BsmtFin Type 2', 'Bsmt Qual', 
        'BsmtFin Type 1', 'Bsmt Cond'
    ]
    df[cat_none_cols] = df[cat_none_cols].fillna('None')
    
    # 3. Garage Yr Blt (Fill with Year Built)
    df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(df['Year Built'])
    
    # 4. Cast categorical number columns to string
    for col in ['MS SubClass', 'Mo Sold', 'Yr Sold']:
        if col in df.columns:
            df[col] = df[col].apply(str)
            
    return df

# Apply to both
train_clean = structural_cleaning(train_df)
test_clean = structural_cleaning(test_df)
```

```python
# Save cleaned data
train_clean.to_csv('Ames_Housing/datasets/train_cleaned.csv', index=False)
test_clean.to_csv('Ames_Housing/datasets/test_cleaned.csv', index=False)
print("Data Cleaning Complete. Saved to datasets/")
```
