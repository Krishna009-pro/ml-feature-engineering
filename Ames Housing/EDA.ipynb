# Ames Housing Dataset — Feature Descriptions

This dataset contains 82 features describing houses in Ames, Iowa. Below is the breakdown of what each feature represents:

### 🏠 1. The Basics (Identification)
*   **Order:** Row number.
*   **PID:** Parcel identification number (unique ID for the land).
*   **MS SubClass:** Type of dwelling (e.g., 1-story, 2-story, split-level).
*   **MS Zoning:** General zoning classification (Residential, Commercial, Agriculture).

### 📏 2. Lot & Location (The Land)
*   **Lot Frontage:** Linear feet of street connected to property.
*   **Lot Area:** Total lot size in square feet.
*   **Street / Alley:** Type of road access (Paved/Gravel).
*   **Lot Shape:** General shape of property (Regular/Irregular).
*   **Land Contour / Slope:** Flatness or steepness of the land.
*   **Neighborhood:** Physical locations within Ames city limits.
*   **Condition 1 & 2:** Proximity to main roads or railroads.
*   **Lot Config:** Lot configuration (e.g., Corner lot, Inside lot).

### 🏗️ 3. Structural Info (The Build)
*   **Bldg Type:** Type of dwelling (Single-family, Townhouse, Duplex).
*   **House Style:** Style of dwelling (1.5 story, 2 story, etc.).
*   **Overall Qual:** Overall material and finish quality (1-10 scale).
*   **Overall Cond:** Overall condition rating (1-10 scale).
*   **Year Built:** Original construction date.
*   **Year Remod/Add:** Remodel date.
*   **Roof Style / Matl:** Type and material of the roof.
*   **Exterior 1st & 2nd:** Exterior covering on the house.
*   **Mas Vnr Type / Area:** Masonry veneer type and area.
*   **Foundation:** Type of foundation (Brick, Concrete, Slab, etc.).

### 🕳️ 4. Basement (Below Ground)
*   **Bsmt Qual:** Height of the basement (Excellent to Poor).
*   **Bsmt Cond:** General condition of the basement.
*   **Bsmt Exposure:** Walkout or garden-level walls.
*   **BsmtFin Type 1 & 2:** Quality of finished basement areas.
*   **BsmtFin SF 1 & 2:** Type 1 & 2 finished square feet.
*   **Bsmt Unf SF:** Unfinished square feet of basement.
*   **Total Bsmt SF:** Total square feet of basement area.

### 🌡️ 5. Utilities & Systems
*   **Heating / Heating QC:** Type of heating and its quality.
*   **Central Air:** Yes/No (Central air conditioning).
*   **Electrical:** Type of electrical system.

### 🛋️ 6. Interior Living Area (The Rooms)
*   **1st Flr SF / 2nd Flr SF:** First and Second floor square feet.
*   **Low Qual Fin SF:** Low quality finished square feet (all floors).
*   **Gr Liv Area:** Above grade (ground) living area square feet.
*   **Full Bath / Half Bath:** Number of bathrooms above ground.
*   **Bsmt Full/Half Bath:** Number of bathrooms in the basement.
*   **Bedroom AbvGr:** Number of bedrooms above basement level.
*   **Kitchen AbvGr:** Number of kitchens.
*   **Kitchen Qual:** Kitchen quality.
*   **TotRms AbvGrd:** Total rooms above ground.
*   **Functional:** Home functionality rating.
*   **Fireplaces / Fireplace Qu:** Number of fireplaces and their quality.

### 🚗 7. Garage
*   **Garage Type:** Garage location (Attached, Detached, Built-in).
*   **Garage Yr Blt:** Year garage was built.
*   **Garage Finish:** Interior finish of the garage.
*   **Garage Cars:** Size of garage in car capacity.
*   **Garage Area:** Size of garage in square feet.
*   **Garage Qual / Cond:** Quality and condition of the garage.

### 🌳 8. Outdoors & Misc
*   **Paved Drive:** Paved driveway (Yes, Partial, No).
*   **Wood Deck SF / Open Porch SF / Enclosed Porch / 3Ssn Porch / Screen Porch:** Square feet of various porch types.
*   **Pool Area / Pool QC:** Pool size and quality.
*   **Fence:** Fence quality.
*   **Misc Feature:** Miscellaneous features (Elevator, Shed, Tennis Court).
*   **Misc Val:** Value of miscellaneous features.

### 💰 9. Sale Information (The Target)
*   **Mo Sold / Yr Sold:** Month and Year the house was sold.
*   **Sale Type:** Type of sale (Warranty Deed, Cash, etc.).
*   **Sale Condition:** Condition of sale (Normal, Abnormality, Partial).
*   **SalePrice:** The price the house sold for (**Target Variable**).

```python
# 1. Neighborhood Impact Analysis
# Calculate Price per Square Foot (Above Grade Living Area)
df['PricePerSF'] = df['SalePrice'] / df['Gr Liv Area']

# Sort neighborhoods by median PricePerSF for a cleaner visual
neighborhood_order = df.groupby('Neighborhood')['PricePerSF'].median().sort_values(ascending=False).index

plt.figure(figsize=(12, 10))
sns.boxplot(x='PricePerSF', y='Neighborhood', data=df, order=neighborhood_order, palette='viridis')
plt.title('Price per Square Foot by Neighborhood (Sorted by Median)', fontsize=15)
plt.xlabel('Price per SqFt ($)')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

### Interpretation: Neighborhood Impact (Price per SqFt)
* **What it tells us:** This is the **"Location Premium."** 
* **Key Insight:** Neighborhoods like *NoRidge* or *StoneBr* don't just have bigger houses; they have a significantly higher cost *per square foot*. This confirms that "Location" is a powerful independent variable. When we build the model, the neighborhood will act as a multiplier for the house's size.

```python
# 2. Temporal Trends Analysis
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Monthly Seasonality (Mean Sale Price)
sns.lineplot(ax=axes[0], x='Mo Sold', y='SalePrice', data=df, estimator='mean', errorbar=None, marker='o', color='royalblue')
axes[0].set_title('Average Sale Price by Month (Seasonality Check)', fontsize=14)
axes[0].set_xticks(range(1, 13))
axes[0].set_xlabel('Month Sold (1=Jan, 12=Dec)')
axes[0].grid(alpha=0.3)

# Yearly Trend (Is the market growing?)
sns.lineplot(ax=axes[1], x='Yr Sold', y='SalePrice', data=df, estimator='mean', errorbar=None, marker='o', color='darkorange')
axes[1].set_title('Average Sale Price by Year (Market Trend)', fontsize=14)
axes[1].set_xticks(df['Yr Sold'].unique())
axes[1].set_xlabel('Year Sold')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### Interpretation: Temporal Trends (Seasonality & Yearly)
* **What it tells us:** The **"Timing of Sale."**
* **Key Insight (Seasonality):** We typically see a peak in Sale Price during the **Summer months (May–July)** and a dip in Winter. This suggests we should treat `Mo Sold` as a seasonal feature.
* **Key Insight (Yearly):** This shows how the market was trending during the 2006–2010 period. It helps identify if we need to adjust for the 2008 housing market crash.

```python
# 3. Interaction: Overall Quality x Neighborhood
# We'll focus on the Top 10 most frequent neighborhoods for clarity in the heatmap
top_neighborhoods = df['Neighborhood'].value_counts().nlargest(10).index
pivot_df = df[df['Neighborhood'].isin(top_neighborhoods)].pivot_table(
    index='Overall Qual', 
    columns='Neighborhood', 
    values='SalePrice', 
    aggfunc='median'
)

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_df, annot=True, fmt=".0f", cmap='YlGnBu', cbar_kws={'label': 'Median Sale Price'})
plt.title('Interaction Analysis: Median SalePrice by Quality & Neighborhood', fontsize=15)
plt.ylabel('Overall Quality (1-10)')
plt.show()
```

### Interpretation: Interaction Heatmap (Quality x Neighborhood)
* **What it tells us:** The **"Multiplier Effect."**
* **Key Insight:** A "High Quality" (Rating 9-10) house in a "Premium" neighborhood has a much higher price jump compared to a "High Quality" house in a budget neighborhood. This tells us that **Quality and Location are NOT independent**—they interact.

# Where We Are & What's Coming Next

---

## 📌 What we understood till now

So the sale price data is little bit skewed on the right side — means most of the houses are on cheaper side and only few are very expensive. This is quite common in housing datasets, nothing to worry about. The simple fix is we will do **log transformation** on the target variable so our model don't get confused by those big values.

Now when we look at what things are actually *affecting* the price, the obvious ones comes up — `Gr Liv Area`, `TotalSF` and `Overall Qual` all have good strong relation with price. But there is some very big houses (like **4000+ sqft**) which are sitting far away from rest of the data, and if we don't handle them carefully they can quietly mess up our model.

The missing data part is little tricky. Most of the `NaN` values are not actually *"missing"* — they just means that house **don't have** that particular feature. Like no garage, no basement, no pool. So we have to treat them like that only, not just fill them randomly.

And also, some features are basically saying same thing two times — like `Garage Cars` and `Garage Area`. If we give both to model it will just create unnecessary noise.

---

## 📌 Things to keep in mind for next step

Few things which will make big difference when we go to next phase:

- **Make new features, don't just use raw columns.**
  Something like `House Age = YrSold − YearBuilt` is much more useful for model then just giving raw year columns directly.

- **Handle missing values properly.**
  For categorical columns put `"None"` and for numerical ones put `0`. Don't just drop the columns — there is useful information hiding inside them.

- **Convert ordinal features into numbers.**
  Ratings like *Excellent → Good → Average → Fair → Poor* already have a order in them. So converting to `5–4–3–2–1` is way better then treating each one as separate category.
